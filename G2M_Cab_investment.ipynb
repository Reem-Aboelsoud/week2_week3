## 1- Data understanding and visualisation, which includes the following analysis

### Profit analysis
The profitability of the two cab companies was compared over the three-year period of 2016, 2017, and 2018. Profit was calculated as the difference between Price Charged and Cost of Trip while other factors were held constant. Analysis and visualizations revealed that the Yellow cab company had higher profits than the Pink cab company over the three-year period. This was particularly evident in the profit ratio visualization, which showed that the Yellow cab company's profit was between 8 to 9 times that of the Pink cab company for the three years analyzed.

### City-wise prift analysis
Across most of the cities in the dataset, the Yellow cab company had a higher number of customers compared to the Pink cab company. This finding was supported by the analysis and visualizations, which showed that the Yellow cab company had higher profits than the Pink cab company over the three-year period in all cities included in the dataset. The profit ratio between the two companies was highest in Dallas, TX with a value of 44.64 and lowest in Sacramento, CA with a value of 1.12. A similar pattern emerged in the number of kilometers traveled by each company, with the Yellow cab company covering a greater distance than the Pink cab company across all cities analyzed.
<br> To obtain a more accurate representation of the total number of users in each city, the analysis assumed that the total number of users includes both Yellow and Pink cab users. The percentage of customers for each company was then calculated as a proportion of the total number of users in each city, and this was visualized to gain further insights. The results of the analysis showed that the Yellow company had a higher percentage of customers in most cities in the dataset, consistent with previous findings.

### Customer analysis
Upon analyzing the data, it was found that the customers' income, age, gender, and distance traveled could be useful factors to consider when deciding which company to invest in. The number of customers traveling a distance between 5 to 40 km is relatively similar between the two companies, but the numbers increase for distances less than five km and more than 40 km. Both companies generate the highest profits from customers with a monthly income between 5000 USD and 25000 USD, while customers earning more than 25000 USD contribute the least to their profits. Customers aged between 25 to 34 years old make the most significant contribution to the two companies' profits, while those over 65 years contribute the least. Additionally, male customers were found to be more numerous than female customers in both companies.
